# 4.9: Intro to Data Visualization with Python

```python
# Importing Libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
# Create a path
path = r'D:\Nov Laptop\Ivan Dimitrov - Data Analyst (CF)\13-06-2023 Instacart Basket Analysis'
```

```python
# Importing Latest Instacart file from Exercise 4.8
df = pd.read_pickle(os.path.join(path, '02 Data','Prepared Data','Orders_Products_Aggregated_4.8.pkl'))
```

```python
# Creating a Bart Chart
df['orders_day_of_week'].value_counts().plot.bar(color =['purple','red','pink','orange','yellow','green','blue'])
```

```python
# Bart chart sorted by index
df['orders_day_of_week'].value_counts().sort_index().plot.bar()
```

```python
# Applying color to the chart
bar = df['orders_day_of_week'].value_counts().plot.bar(color =['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
```

```python
bar.figure
```

```python
# Exporting the Bar chart
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

```python
# Creating a Histogram
df['prices'].plot.hist(bins=25)
```

```python
# Creating a Scatterplot
sns.scatterplot(x = 'prices', y = 'prices',data = df)
```

```python
pd.set_option('display.max_columns', None)
```
