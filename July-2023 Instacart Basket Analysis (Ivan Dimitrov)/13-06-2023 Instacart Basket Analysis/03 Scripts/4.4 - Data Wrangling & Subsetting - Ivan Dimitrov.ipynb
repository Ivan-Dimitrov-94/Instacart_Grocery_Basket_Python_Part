### Step 6: The sales team in your client’s organization wants to know more about breakfast item sales. Create a subset containing only the required information.

```python
# From the DD we know that Nr.14 is "breakfast", so I made a subset for Nr.14 only and sorted it according to prices descending.
df_breakfast = df_prods.loc[df_prods['department_id'].isin([14])]
df_breakfast.sort_values('prices', ascending=False)
```

### Step 7: They’d also like to see details about products that customers might use to throw dinner parties. Your task is to find all observations from the entire dataframe that include items from the following departments: alcohol, deli, beverages, and meat/seafood. You’ll need to present this subset to your client.

```python
# From the DD we know the departments for the respective categories:
# Alcohol-5 ; Deli-20 ; Beverages-7 ; Meat/Seafood-12
# So we use the same function again with "isin([])"
df_breakfast = df_prods.loc[df_prods['department_id'].isin([5,20,7,12])]
df_breakfast.head(20).sort_values('prices', ascending=False)
#As we can see Meat/Seafood and beverages prevail
```
